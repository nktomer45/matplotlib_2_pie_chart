# Ploting histograms

```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
ml_students_age=np.random.randint(15,50,(100))
java_students_age=np.random.randint(15,50,(100))
```

```python
display(ml_students_age,java_students_age)
```

```python
plt.hist(java_students_age)
plt.xlabel("student catgierous",fontsize=20)
plt.ylabel("age",fontsize=20)
plt.title("students age data",fontsize=24)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_students_age, bins,rwidth=0.8,histtype="bar",orientation='horizontal')
plt.hist(java_students_age, bins,rwidth=0.8,histtype="bar",orientation='horizontal',color="pink")
plt.xlabel("student catgierous",fontsize=20)
plt.ylabel("age",fontsize=20)
plt.title("students age data",fontsize=24)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_students_age, bins,rwidth=0.8,histtype="bar",orientation='horizontal',color="pink")
plt.xlabel("student catgierous",fontsize=20)
plt.ylabel("age",fontsize=20)
plt.title("students age data",fontsize=24)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_students_age, bins,rwidth=0.8,histtype="bar",orientation='vertical',color="pink",label="ml_std")
plt.hist(java_students_age, bins,rwidth=0.8,histtype="bar",orientation='vertical',color="y",label="java_std")
plt.xlabel("student catgierous",fontsize=20)
plt.ylabel("age",fontsize=20)
plt.title("students age data",fontsize=24)
plt.legend(fontsize=10)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,java_students_age] ,bins,rwidth=0.8,histtype="bar",orientation='vertical',color=["pink","yellow"],label=["ml_std","java_std"])

plt.xlabel("student catgierous",fontsize=20)
plt.ylabel("age",fontsize=20)
plt.title("students age data",fontsize=24)
plt.legend(fontsize=10)
plt.show()
```
