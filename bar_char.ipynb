# ploting bar chart

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["python","ML","java","sql","c"]
class1_student=[12,23,20,34,50]
class2_student=[10,21,23,30,40]
class3_student=[14,24,30,43,39]
```

```python
plt.bar(classes,class1_student)
```

```python
plt.barh(classes,class1_student)
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color='yellow',edgecolor='green',linewidth=3,alpha=.5,linestyle="--"
       ,label="classs1_student")
# alpha is used for transparence of edge color
```

# ploting bar chart 

```python
classes=["python","ML","java","sql","c"]
class1_student=[12,23,20,34,50]
class2_student=[10,21,23,30,40]
class3_student=[14,24,30,43,39]
```

```python
style.use("ggplot")
plt.figure(figsize=(8,4))
plt.grid(color='red',linestyle='--')
plt.bar(classes,class1_student,width=0.2,align="edge",color='yellow',edgecolor='green',linewidth=3,alpha=.5,linestyle="--"
       ,label="classs1_student")
```

```python
style.use("ggplot")
plt.figure(figsize=(8,4))
plt.grid(color='red',linestyle='--')
plt.barh(classes,class1_student,align="edge",color='yellow',
         edgecolor='green',linewidth=3,alpha=.5,linestyle="--"
       ,label="classs1_student")
```

```python
classes=["python","ML","java","sql","c"]
class1_student=[12,23,20,34,50]
class2_student=[10,21,23,30,40]
class3_student=[34,26,30,43,39]

style.use("ggplot")
plt.figure(figsize=(8,4))
classes_index=np.arange(len(classes))

width=.2

#plt.grid(color='red',linestyle='--')
plt.bar(classes_index +  width,class1_student,width,color='yellow',
       label="classs1_student")
plt.bar(classes_index + width+width,class2_student,width,color='green',
       label="classs2_student")

plt.bar(classes_index + width+width+width,class2_student,width,color='b',
       label="classs2_student") 

plt.title("categeious of student",fontsize=10)
plt.xlabel("class",fontsize=10)
plt.ylabel("student",fontsize=10)
plt.xticks(classes_index +  width,classes)
plt.show()
```
